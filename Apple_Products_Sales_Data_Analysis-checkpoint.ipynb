## Checking for Outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Columns to check for outliers
cols_to_check = ['quantity', 'price', 'revenue']

# Copy dataset for analysis/plots
Apple_sales_plot = Apple_products_sales_data.copy()

# Function to detect and cap outliers
def cap_outliers(df, col):
    # 1st and 99th percentiles
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    
    # Detect outliers
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers detected")
    
    # Cap outliers
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
    
    return df

# Apply capping to all numeric columns
for col in cols_to_check:
    Apple_sales_plot = cap_outliers(Apple_sales_plot, col)

# Optional: visualize capped data
for col in cols_to_check:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=Apple_sales_plot[col])
    plt.title(f'Boxplot for {col} (Capped)')
    plt.show()
```

## 3:What are the top 10 most selling products ?

## 4.Which products generate the more revenue among all products?

```python
Apple_sales_plot['revenue_per_unit'] = Apple_sales_plot['revenue'] / Apple_sales_plot['quantity']
top_revenue_products = Apple_sales_plot.groupby('product_name')['revenue_per_unit'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_revenue_products.values, y=top_revenue_products.index, palette='cividis')
plt.title('Top 10 Products by Revenue per Unit')
plt.xlabel('Revenue per Unit')
plt.ylabel('Product')
plt.show()
plt.tight_layout()
```

## Insight:

+ These may sell less frequently but generate more revenue per sale.
